#### Step 1:  State null and alternative hypotheses

Null Hypotheses- $H_0$:  The stone surface Brinell's hardness index of at least 150(μ<= 150)

Alternate Hypotheses- $H_A$: The stone surface  Brinell's hardness index of Greater than 150 (μ> 150)




#### Step 2: significance level
α  = 0.05, and the population standard deviation is not known.


#### Step 3:  Hypothesis test and compute the p-value

```python
#The population standard deviation is unknown. So we use the t distribution and the  tSTAT  test statistic.(One Tailed Test)
t_statistics, p_value1 = stats.ttest_1samp(df['Unpolished '], 150)
```

```python
#For one tailed T- Test, P value needs to be divide by two
p_value=p_value1/2
```

```python
p_value
```

```python
print("t_statistics={} and p_value = {}".format(round(t_statistics,3), round(p_value,7)))
```

#### Step 4: Decide to reject or accept null hypothesis

```python
# Level of significance
alpha=0.05
if (p_value<alpha):
   print('We have evidence to reject the null hypothesis since p value < Level of significance')
else:
   print('We have no evidence to reject the null hypothesis since p value > Level of significance')
```

Conclusion: So at 95% confidence level, there is sufficient evidence to prove that The stone surface Brinell's hardness index of Greater than 150 (μ> 150) and unpolished stones may not be suitable for printing.

For testing of normality, Shapiro-Wilk’s test is applied to the response to check the Normality

* $H_0$: Response follows a normal distribution against

* $H_A$: Response does not  follow a normal distribution

```python
#Aussmption 1: Normality-  Shapiro-Wilk’s test
from scipy import stats
w, p_value = stats.shapiro(df_alloy1['Response'])
print("W = {}".format(w), "p_value = {}".format(p_value))
```

```python
#Condition to reject the Null Hypothsis
alpha=0.05
if (p_value<alpha):
   print("Reject Null Hypothesis")
else:
   print(" Fail to Reject Null Hypothesis")
```

Since p-value of the test is very small, we reject the null hypothesis means that the Response is not normally distributed.

For alloy 1 Response is not normally distributed

Levene test to assumption two : If the variances are significantly different,

𝐻0: 𝜎1 = 𝜎2 = 𝜎3 =𝜎4=𝜎5

𝐻𝑎: At least one variance is different from the rest.

```python
##Aussmption 2: Homogeneity of Variance
from scipy.stats import levene
statistic, p_value = levene( df_alloy1['Response'][df_alloy1['Dentist']=="Dentist_1"],
                                 df_alloy1['Response'][df_alloy1['Dentist']=="Dentist_2"],
                                 df_alloy1['Response'][df_alloy1['Dentist']=="Dentist_3"],
                                 df_alloy1['Response'][df_alloy1['Dentist']=="Dentist_4"],
                                 df_alloy1['Response'][df_alloy1['Dentist']=="Dentist_5"])
# find the p-value
print("statistic = {}".format(statistic), "p_value = {}".format(p_value))
```

```python
#Condition to reject the Null Hypothsis
alpha=0.05
if (p_value<alpha):
   print("Reject Null Hypothesis")
else:
   print(" Fail to Reject Null Hypothesis")
```

Since the p-value is large, we fail to reject the null hypothesis of homogeneity of variances.

Hence, The variance of the response variable is the same for all of the populations of Dentist Variable

Compute the p-value

```python
#Solution for Hypothesis test and compute the p-value
from scipy.stats import stats
from statsmodels.formula.api import ols
import statsmodels.api as sm
mod = ols('Response ~ Dentist', data = df_alloy1).fit()
aov_tbl = sm.stats.anova_lm(mod, type = 1)
print(aov_tbl)
```

```python
alpha=0.05
p_value=0.116567
if (p_value<alpha):
   print("Reject the null hypothesis. There is significant difference between means.")
else:
    print("Fail to reject the null hypothesis. There is no significant")
```
